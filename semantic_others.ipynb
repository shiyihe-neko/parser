```python
import re
import json
import pandas as pd
import yaml, toml, xmltodict
from typing import Any, Dict, Set, Tuple
from zss import Node, simple_distance
from tree_sitter_language_pack import get_parser

# ----------------------------------------------------------------------------
# 1. 三阶段解析函数 (可选 Tree-Sitter)
# ----------------------------------------------------------------------------
parser_yaml = get_parser("yaml")
parser_toml = get_parser("toml")
parser_xml  = get_parser("xml")

def node_to_dict(node, src: bytes):
    # 按需将 tree-sitter AST 转为 Python 对象，略同前文
    if node.type in ("mapping","flow_mapping","object","document","table","inline_table"):
        d = {}
        for pair in node.named_children:
            if pair.type.startswith(("pair","map_pair","pair_item","table_pair")):
                key = pair.child_by_field_name("key") or pair.named_children[0]
                val = pair.child_by_field_name("value") or pair.named_children[-1]
                d[node_to_dict(key, src)] = node_to_dict(val, src)
        return d
    if node.type in ("sequence","flow_sequence","array","array_elements"):
        lst = []
        for itm in node.named_children:
            child = itm.named_children[0] if itm.named_children else itm
            lst.append(node_to_dict(child, src))
        return lst
    txt = src[node.start_byte:node.end_byte].decode("utf8")
    if node.type in ("string","quoted_scalar","text"): return txt.strip('"\'')
    if node.type in ("integer","float","number"):
        try: return int(txt)
        except: return float(txt) if '.' in txt else txt
    if node.type == "boolean": return txt.lower() == 'true'
    if node.type == "null":    return None
    return txt

# 解析主函数
def parse_code(code: str, fmt: str, use_tree_sitter: bool = True) -> Tuple[Dict[str,Any], str]:
    fmt = fmt.lower()
    # XML 优先 xmltodict
    if fmt == "xml":
        txt = re.sub(r"<\?xml[^>]+\?>", "", code)
        try:
            d = xmltodict.parse(txt)
            if isinstance(d, dict) and len(d) == 1:
                d = next(iter(d.values()))
            return d, 'xmltodict'
        except:
            pass
        if use_tree_sitter:
            try:
                tree = parser_xml.parse(txt.encode('utf8'))
                d = node_to_dict(tree.root_node, txt.encode('utf8'))
                if d: return d, 'tree-sitter'
            except: pass
        out = {}
        for m in re.finditer(r"([\w\-]+)\s*[:=]\s*([^,\n]+)", txt):
            out[m.group(1)] = m.group(2).strip()
        return out, 'regex'

    # YAML
    if fmt == "yaml":
        if use_tree_sitter:
            try:
                tree = parser_yaml.parse(code.encode('utf8'))
                d = node_to_dict(tree.root_node, code.encode('utf8'))
                if d: return d, 'tree-sitter'
            except: pass
        try:
            d = yaml.safe_load(code)
            if d: return d, 'library'
        except: pass
        out = {}
        for m in re.finditer(r"([\w\-]+)\s*[:=]\s*([^,\n]+)", code):
            out[m.group(1)] = m.group(2).strip()
        return out, 'regex'

    # TOML
    if fmt == "toml":
        if use_tree_sitter:
            try:
                tree = parser_toml.parse(code.encode('utf8'))
                d = node_to_dict(tree.root_node, code.encode('utf8'))
                if d: return d, 'tree-sitter'
            except: pass
        try:
            d = toml.loads(code)
            if d: return d, 'library'
        except: pass
        out = {}
        for m in re.finditer(r"([\w\-]+)\s*[:=]\s*([^,\n]+)", code):
            out[m.group(1)] = m.group(2).strip()
        return out, 'regex'

    # fallback
    out = {}
    for m in re.finditer(r"([\w\-]+)\s*[:=]\s*([^,\n]+)", code):
        out[m.group(1)] = m.group(2).strip()
    return out, 'regex'

# ----------------------------------------------------------------------------
# 2. 指标计算
# ----------------------------------------------------------------------------
def flatten(obj: Any, parent_key: str = "") -> Dict[str, Any]:
    flat = {}
    if isinstance(obj, dict):
        for k,v in obj.items(): flat.update(flatten(v, f"{parent_key}.{k}" if parent_key else k))
    elif isinstance(obj, list):
        for i,v in enumerate(obj): flat.update(flatten(v, f"{parent_key}[{i}]") )
    else: flat[parent_key] = obj
    return flat

def prf(pred: Set, true: Set) -> Tuple[float,float,float]:
    tp=len(pred&true); fp=len(pred-true); fn=len(true-pred)
    p=tp/(tp+fp) if tp+fp else 0; r=tp/(tp+fn) if tp+fn else 0
    return p, r, (2*p*r/(p+r) if p+r else 0)

def dict_metrics(u: Dict[str,Any], g: Dict[str,Any]) -> Dict[str,float]:
    uf=flatten(u); gf=flatten(g)
    kp,kr,kf=prf(set(uf), set(gf))
    shared=set(uf)&set(gf)
    matches={k for k in shared if uf[k]==gf[k]}
    vp=len(matches)/len(shared) if shared else 0
    vr=len(matches)/len(gf)     if gf     else 0
    vf=(2*vp*vr/(vp+vr) if vp+vr else 0)
    ivp,ivr,ivf=prf(set(uf.items()), set(gf.items()))
    return {"keys_precision":kp, "keys_recall":kr, "keys_f1":kf,
            "values_precision":vp, "values_recall":vr, "values_f1":vf,
            "items_precision":ivp, "items_recall":ivr, "items_f1":ivf}

def dict_to_zss(node: Any, label: str="root") -> Node:
    if not isinstance(node,(dict,list)): return Node(f"{label}:{node}")
    kids=[]
    if isinstance(node, dict):
        for k in sorted(node): kids.append(dict_to_zss(node[k], label=k))
    else:
        for i,v in enumerate(node): kids.append(dict_to_zss(v, label=f"{label}[{i}]") )
    return Node(label, kids)

def tree_metrics(u: Dict[str,Any], g: Dict[str,Any]) -> Dict[str,float]:
    n1=dict_to_zss(u,"root"); n2=dict_to_zss(g,"root")
    dist=simple_distance(n1,n2)
    def cnt(n): return 1+sum(cnt(c) for c in n.children)
    m=max(cnt(n1),cnt(n2)) or 1
    return {"tree_distance":dist, "normalized_distance":dist/m, "tree_similarity":1-dist/m}


# ----------------------------------------------------------------------------
# 3. 流水线 evaluate_df
# ----------------------------------------------------------------------------




# def evaluate_df(df_in: pd.DataFrame, use_tree_sitter: bool = True) -> pd.DataFrame:
#     df=df_in.copy()
#     # 解析
#     parsed=[]; methods=[]
#     for _,row in df.iterrows():
#         d,m=parse_code(row['code'],row['format'], use_tree_sitter)
#         parsed.append(d); methods.append(m)
#     df['parsed_code']=parsed
#     df['parsed_method']=methods
#     # 转 JSON->dict 保证干净

#     df['parsed_code_json']=df['parsed_code'].apply(lambda d: json.dumps(d, ensure_ascii=False, sort_keys=True))
#     df['parsed_code_dict']=df['parsed_code_json'].apply(json.loads)
#     kv=df['parsed_code_dict'].apply(lambda u: dict_metrics(u,gold_dict))
#     tr=df['parsed_code_dict'].apply(lambda u: tree_metrics(u,gold_dict))
#     kv_df=pd.DataFrame(kv.tolist(), index=df.index)
#     tr_df=pd.DataFrame(tr.tolist(), index=df.index)
#     # 合并并保留原 columns
#     return pd.concat([df[['participantId','format','code','parsed_code','parsed_code_dict','parsed_method']], kv_df, tr_df], axis=1)
# # 修改evaluate_df函数，为XML格式数据添加特殊处理
# # 其他逻辑保持不变


def evaluate_df(
    df_in: pd.DataFrame,
    gold_dict: dict,
    use_tree_sitter: bool = True
) -> pd.DataFrame:
    """
    Args:
      df_in:            包含 'participantId','format','code' 三列的输入 DataFrame
      gold_dict:        真值 dict，用于 dict_metrics 和 tree_metrics
      use_tree_sitter:  是否启用 Tree-sitter
    
    Returns:
      包含解析结果和各项指标的 DataFrame
    """
    df = df_in.copy()

    # 1) 解析阶段
    parsed, methods = [], []
    for _, row in df.iterrows():
        d, m = parse_code(row['code'], row['format'], use_tree_sitter)
        parsed.append(d)
        methods.append(m)
    df['parsed_code']   = parsed
    df['parsed_method'] = methods

    # 3) 计算指标
    # dict_metrics 返回 (key_prf, value_prf, kv_prf)
    df['parsed_code_json']=df['parsed_code'].apply(lambda d: json.dumps(d, ensure_ascii=False, sort_keys=True))
    df['parsed_code_dict']=df['parsed_code_json'].apply(json.loads)
    kv=df['parsed_code_dict'].apply(lambda u: dict_metrics(u,gold_dict))
    tr=df['parsed_code_dict'].apply(lambda u: tree_metrics(u,gold_dict))
    kv_df=pd.DataFrame(kv.tolist(), index=df.index)
    tr_df=pd.DataFrame(tr.tolist(), index=df.index)
    # 合并并保留原 columns
    return pd.concat([df[['participantId','format','code','parsed_code','parsed_code_dict','parsed_method']], kv_df, tr_df], axis=1)


import os
import pandas as pd

def run_evaluation_pipeline(
    data_df: pd.DataFrame,
    gold_dict: dict,
    output_dir: str,
    file_prefix: str,
    use_tree_sitter: bool = True,
    verbose: bool = False
) -> pd.DataFrame:
    """
    通用的解析 + 指标计算 + 保存流程（基于 evaluate_df）

    Args:
      data_df:         待处理的 DataFrame，需包含 'participantId','format','code' 列
      gold_dict:       真值 dict，用于 dict_metrics & tree_metrics
      output_dir:      结果保存目录
      file_prefix:     输出文件名前缀，如 'writing_tabular_toml'
      use_tree_sitter: 是否启用 tree-sitter 解析
      verbose:         是否打印进度信息

    Returns:
      合并了解析结果和指标的 DataFrame，同时写入 CSV 文件
    """
    # 1) 调用 evaluate_df 完成解析与指标计算
    result_df = evaluate_df(
        df_in=data_df,
        gold_dict=gold_dict,
        use_tree_sitter=use_tree_sitter
    )


    # 2) 构造文件名
    filename = f"{file_prefix}_result.csv"
    os.makedirs(output_dir, exist_ok=True)
    out_path = os.path.join(output_dir, filename)

    # 3) 保存
    result_df.to_csv(out_path, index=False, encoding='utf-8')
    if verbose:
        print(f"Saved evaluation results to {out_path!r}")

    return result_df
```

```python
import re
import json
import pandas as pd
import xmltodict
from typing import Any, Dict, Set, Tuple
from zss import Node, simple_distance
from tree_sitter_language_pack import get_parser
import xml.etree.ElementTree as ET


# ----------------------------------------------------------------------------
# 0. XML 预处理：去掉声明/注释，修正标签名，wrap 根
# ----------------------------------------------------------------------------
def sanitize_xml_fragment(x: str) -> str:
    # 抹掉 XML 声明、DOCTYPE、注释、PI
    x = re.sub(r"<\?xml[^>]*\?>", "", x)
    x = re.sub(r"<!DOCTYPE[^>]*>", "", x)
    x = re.sub(r"<!--.*?-->", "", x, flags=re.DOTALL)
    # 把非法标签名中间的空格或其他非字母数字字符替换成下划线
    def fix_tag(m):
        slash, name = m.groups()
        name_fixed = re.sub(r"[^0-9A-Za-z]+", "_", name)
        return f"<{slash}{name_fixed}>"
    x = re.sub(r"<\s*(/?)\s*([\w\- ]+?)\s*>", fix_tag, x)
    # wrap
    return f"<root>{x}</root>"


parser_xml = get_parser("xml")


def xml_regex_to_dict(xml: str) -> dict:
    """
    1) 递归解析任意嵌套的 <tag>…</tag> 结构
    2) 如果递归结果为空，再全局匹配一次 <tag>value</tag> ，聚合同名为列表
    """
    def _parse(seg: str) -> dict:
        result = {}
        # 非贪婪匹配最外层标签
        pattern = re.compile(r"<(?P<tag>[A-Za-z0-9_:-]+)>(?P<inner>.*?)</(?P=tag)>", re.DOTALL)
        for m in pattern.finditer(seg):
            tag   = m.group("tag")
            inner = m.group("inner").strip()
            # 如果里面还有子标签，就递归
            if pattern.search(inner):
                val = _parse(inner)
            else:
                val = re.sub(r"<[^>]+>", "", inner).strip()
            # 聚合同名标签为列表
            if tag in result:
                if isinstance(result[tag], list):
                    result[tag].append(val)
                else:
                    result[tag] = [result[tag], val]
            else:
                result[tag] = val
        return result

    # 1) 先试递归嵌套
    nested = _parse(xml)
    if nested:
        return nested

    # 2) 如果真的没拿到任何标签，再做全局 KV 提取
    flat: dict = {}
    for tag, val in re.findall(r"<([A-Za-z0-9_:-]+)>\s*([^<]+?)\s*</\1>", xml):
        v = val.strip()
        if tag in flat:
            if isinstance(flat[tag], list):
                flat[tag].append(v)
            else:
                flat[tag] = [flat[tag], v]
        else:
            flat[tag] = v
    return flat


def parse_code(code: str, fmt: str, use_tree_sitter: bool = True) -> (dict, str):
    fmt = fmt.lower().strip()

    if fmt == "xml":
        # 1. sanitize & wrap 根
        txt = re.sub(r"<\?xml[^>]*\?>", "", code)
        txt = re.sub(r"<!DOCTYPE[^>]*>", "", txt)
        txt = re.sub(r"<!--.*?-->", "", txt, flags=re.DOTALL)
        wrapped = f"<root>{txt}</root>"

        # 2. 尝试 Tree-Sitter
        if use_tree_sitter:
            try:
                tree = parser_xml.parse(wrapped.encode("utf8"))
                d = node_to_dict(tree.root_node, wrapped.encode("utf8"))
                result = (d.get("root") if isinstance(d, dict) else None) or d
                if isinstance(result, dict) and result:
                    return result, "tree-sitter"
            except:
                pass

        # 3. 尝试 xmltodict
        try:
            d = xmltodict.parse(wrapped)
            result = (d.get("root") if isinstance(d, dict) else None) or d
            if isinstance(result, dict) and result:
                return result, "xmltodict"
        except:
            pass

        # 4. 尝试 xml.etree 回退
        try:
            root = ET.fromstring(wrapped)
            # 跳过最外层 root
            d = {child.tag: etree_to_dict(child) for child in root}
            if d:
                return d, "etree"
        except:
            pass

        # 5. 最后正则兜底
        # fallback = xml_regex_to_dict(wrapped)
        # return fallback, "regex"
        fallback = xml_regex_to_dict(wrapped)
        return fallback, "regex"

    # —— 你的 YAML/TOML/JSON 分支保持不变 —— #
    # 这里示例最简正则回退
    out = {m.group(1): m.group(2).strip()
           for m in re.finditer(r"([\w\-]+)\s*[:=]\s*([^,\n]+)", code)}
    return out, "regex"

# ----------------------------------------------------------------------------
# 1. 三阶段解析函数 (可选 Tree-Sitter)
# ----------------------------------------------------------------------------
parser_xml = get_parser("xml")

def node_to_dict(node, src: bytes) -> Any:
    # 同前：完整实现略...
    if node.type in ("mapping","flow_mapping","object","document","table","inline_table"):
        d = {}
        for pair in node.named_children:
            if pair.type.startswith(("pair","map_pair","pair_item","table_pair")):
                key_node = pair.child_by_field_name("key") or pair.named_children[0]
                val_node = pair.child_by_field_name("value") or pair.named_children[-1]
                d[node_to_dict(key_node, src)] = node_to_dict(val_node, src)
        return d
    if node.type in ("sequence","flow_sequence","array","array_elements"):
        return [node_to_dict(c if not c.named_children else c.named_children[0], src)
                for c in node.named_children]
    txt = src[node.start_byte:node.end_byte].decode("utf8")
    if node.type in ("string","quoted_scalar","text"): return txt.strip('"\'')
    if node.type in ("integer","float","number"):
        try: return int(txt)
        except: return float(txt) if '.' in txt else txt
    if node.type == "boolean": return txt.lower()=='true'
    if node.type == "null":    return None
    return txt

# ----------------------------------------------------------------------------
# 2. 扁平化 & PRF
# ----------------------------------------------------------------------------
def flatten(obj: Any, parent_key: str = "") -> Dict[str, Any]:
    flat = {}
    if isinstance(obj, dict):
        for k,v in obj.items():
            nk = f"{parent_key}.{k}" if parent_key else k
            flat.update(flatten(v, nk))
    elif isinstance(obj, list):
        for i,v in enumerate(obj):
            nk = f"{parent_key}[{i}]"
            flat.update(flatten(v, nk))
    else:
        flat[parent_key] = obj
    return flat

def prf(pred: Set, true: Set) -> Tuple[float,float,float]:
    tp = len(pred & true)
    fp = len(pred - true)
    fn = len(true - pred)
    p  = tp / (tp+fp) if tp+fp else 0.0
    r  = tp / (tp+fn) if tp+fn else 0.0
    f1 = 2*p*r/(p+r) if p+r else 0.0
    return p, r, f1

def dict_metrics(u: Dict[str,Any], g: Dict[str,Any]) -> Dict[str,float]:
    uf = flatten(u); gf = flatten(g)
    kp,kr,kf = prf(set(uf), set(gf))
    shared   = set(uf)&set(gf)
    matches  = {k for k in shared if uf[k]==gf[k]}
    vp = len(matches)/len(shared) if shared else 0.0
    vr = len(matches)/len(gf)     if gf     else 0.0
    vf = 2*vp*vr/(vp+vr) if vp+vr else 0.0
    ivp,ivr,ivf = prf(set(uf.items()), set(gf.items()))
    return {
      "keys_precision":   kp, "keys_recall":   kr, "keys_f1":   kf,
      "values_precision": vp, "values_recall": vr, "values_f1": vf,
      "items_precision":  ivp,"items_recall":  ivr,"items_f1": ivf
    }

# ----------------------------------------------------------------------------
# 3. 树编辑距离
# ----------------------------------------------------------------------------
def dict_to_zss(node: Any, label: str="root") -> Node:
    if not isinstance(node,(dict,list)):
        return Node(f"{label}:{node}")
    kids = []
    if isinstance(node, dict):
        for k in sorted(node):
            kids.append(dict_to_zss(node[k], label=k))
    else:
        for i,v in enumerate(node):
            kids.append(dict_to_zss(v, label=f"{label}[{i}]"))
    return Node(label, kids)

def tree_metrics(u: Dict[str,Any], g: Dict[str,Any]) -> Dict[str,float]:
    n1 = dict_to_zss(u,"root"); n2 = dict_to_zss(g,"root")
    dist = simple_distance(n1, n2)
    def cnt(n): return 1 + sum(cnt(c) for c in n.children)
    m = max(cnt(n1), cnt(n2)) or 1
    return {
      "tree_distance": dist,
      "normalized_distance": dist/m,
      "tree_similarity": 1 - dist/m
    }

def unwrap_root(obj: dict, method: str) -> dict:
    # 只有 regex 回退，且顶层只有一个 root 键时，才去除它
    if method == "regex" and isinstance(obj, dict) and "root" in obj and len(obj) == 1:
        inner = obj["root"]
        # 确保拿到的也是 dict
        return inner if isinstance(inner, dict) else {"#text": inner}
    return obj
# ----------------------------------------------------------------------------
# 4. 流水线 evaluate_df
# ----------------------------------------------------------------------------
# gold_json = '''{
#   "version": "4.4.7",
#   "repository": {"author":{"middle name":"","IsMember":true},"type":"git"},
#   "engines":   {"pnpm":">=8"},
#   "keywords":  ["canvas","responsive","charts"]
# }'''

gold_json = '''{
  "patients": [
    {
      "name": "John",
      "tests": [
        {
          "id": 1,
          "result": "Normal"
        },
        {
          "id": 2,
          "result": "Elevated Cholesterol"
        }
      ]
    },
    {
      "name": "Michael",
      "tests": [
        {
          "id": 1,
          "result": "Normal"
        }
      ],
      "treatment": "Regular Monitoring"
    }
  ]
}'''
gold_dict = json.loads(gold_json)

def evaluate_df_xml(df_in: pd.DataFrame, use_tree_sitter: bool = True) -> pd.DataFrame:
    df = df_in.copy()
    # parse
    parsed, methods = [], []
    for _, row in df.iterrows():
        d,m = parse_code(row["code"], row["format"], use_tree_sitter)
        parsed.append(d); methods.append(m)
    df["parsed_code"]   = parsed
    df["parsed_method"] = methods
    df["parsed_code"] = df.apply(
         lambda r: unwrap_root(r["parsed_code"], r["parsed_method"]), axis=1
     )

    # serialize & clean
    df["parsed_code_json"] = df["parsed_code"].apply(
        lambda d: json.dumps(d, ensure_ascii=False, sort_keys=True)
    )
    df["parsed_code_dict"] = df["parsed_code_json"].apply(json.loads)
    # metrics
    kv = df["parsed_code_dict"].apply(lambda u: dict_metrics(u, gold_dict))
    tr = df["parsed_code_dict"].apply(lambda u: tree_metrics(u,     gold_dict))
    kv_df = pd.DataFrame(kv.tolist(), index=df.index)
    tr_df = pd.DataFrame(tr.tolist(), index=df.index)
    # 合并
    return pd.concat([
      df[[
        "participantId","format","code",
        "parsed_code","parsed_code_dict","parsed_method","parsed_code_json"
      ]],
      kv_df, tr_df
    ], axis=1)


import json
import pandas as pd

def evaluate_df_xml(
    df_in: pd.DataFrame,
    gold_dict: dict,
    use_tree_sitter: bool = True
) -> pd.DataFrame:

    df = df_in.copy()

    # 1) 解析阶段
    parsed, methods = [], []
    for _, row in df.iterrows():
        d, m = parse_code(row["code"], row["format"], use_tree_sitter)
        parsed.append(d)
        methods.append(m)
    df["parsed_code"]   = parsed
    df["parsed_method"] = methods

    # 2) 如果是 XML 且有包裹 <root>，去掉外层
    df["parsed_code"] = df.apply(
        lambda r: unwrap_root(r["parsed_code"], r["parsed_method"]),
        axis=1
    )

    # 3) 序列化 & 清洗
    df["parsed_code_json"] = df["parsed_code"].apply(
        lambda d: json.dumps(d, ensure_ascii=False, sort_keys=True)
    )
    df["parsed_code_dict"] = df["parsed_code_json"].apply(json.loads)

    # 4) 计算指标
    # dict_metrics 返回 {'key':(...), 'value':(...), 'kv':(...)}
    kv_metrics = df["parsed_code_dict"].apply(lambda u: dict_metrics(u, gold_dict))
    kv_df = pd.DataFrame(kv_metrics.tolist(), index=df.index)

    # tree_metrics 返回 {'ted','normalized_ted','similarity'}
    tree_metrics_list = df["parsed_code_dict"].apply(lambda u: tree_metrics(u, gold_dict))
    tree_df = pd.DataFrame(tree_metrics_list.tolist(), index=df.index)

    # 5) 合并结果
    result_df = pd.concat([
        df[[
            "participantId", "format", "code",
            "parsed_code", "parsed_code_dict",
            "parsed_method", "parsed_code_json"
        ]],
        kv_df,
        tree_df
    ], axis=1)

    return result_df


import os
import pandas as pd

def run_xml_pipeline(
    data_df: pd.DataFrame,
    gold_dict: dict,
    output_dir: str,
    file_prefix: str,
    use_tree_sitter: bool = True,
    verbose: bool = False
) -> pd.DataFrame:
    """
    通用的 XML 解析 + 指标计算 + 保存流程（基于 evaluate_df_xml）

    Args:
      data_df:         待处理的 DataFrame，需包含 'participantId','format','code' 列
      gold_dict:       真值 dict，用于 dict_metrics & tree_metrics
      output_dir:      结果保存目录
      file_prefix:     输出文件名前缀（不含后缀部分）
      use_tree_sitter: 是否启用 tree-sitter 解析
      verbose:         是否打印进度信息

    Returns:
      合并了解析结果和指标的 DataFrame，同时写入 CSV 文件
    """
    # 1) 调用 evaluate_df_xml 完成解析与指标计算
    result_df = evaluate_df_xml(
        df_in=data_df,
        gold_dict=gold_dict,
        use_tree_sitter=use_tree_sitter
    )

    # 2) 构造文件名并保存
    filename = f"{file_prefix}_result.csv"
    os.makedirs(output_dir, exist_ok=True)
    out_path = os.path.join(output_dir, filename)

    result_df.to_csv(out_path, index=False, encoding='utf-8')
    if verbose:
        print(f"Saved XML evaluation results to {out_path!r}")

    return result_df


with open('ground_truth_writing_tabular.json', 'r', encoding='utf-8') as f:
    gold = json.load(f)

# 调用 pipeline
df_xml_1 = run_xml_pipeline(
    data_df= writing_tabular_xml,
    gold_dict= gold,
    output_dir="/Users/shiyi.he/Desktop/PARSER",
    file_prefix="dataset1_xml",
    use_tree_sitter=True,
    verbose=False
)
```
